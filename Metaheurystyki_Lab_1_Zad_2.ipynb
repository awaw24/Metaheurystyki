Moja implementacja używa generatora LCG do generowania liczb z rozkładu jednostajnego na [0,1].

Oblicza średnią oraz odchylenie standardowe wygenerowanych próbek.

Rysuje histogram rozkładu.

Wyświetla sprawozdanie porównujące wartości uzyskane z teoretycznymi dla rozkładu jednostajnego oblicza:
- średnią
- odchylenie standardowe

```python
import numpy as np
import matplotlib.pyplot as plt

def lcg(a, c, m, seed, n):
    """
    Klasyczny generator liniowy (LCG):
      X(n+1) = (a * X(n) + c) mod m
    Liczby są normalizowane do przedziału [0,1] przez podzielenie przez m.
    """
    X = [seed]
    for i in range(1, n):
        X.append((a * X[-1] + c) % m)
    normalized = [x / m for x in X]
    return normalized

def compute_stats(random_numbers):
    """
    Obliczanie średniej oraz odchylenia standardowego dla listy liczb.
    """
    mean = np.mean(random_numbers)
    std = np.std(random_numbers)
    return mean, std

if __name__ == "__main__":
    # Parametry generatora LCG
    a = 1103515245
    c = 12345
    m = 2**31
    seed = 42
    n = 10000  # liczba generowanych próbek

    # Generacja liczb losowych
    random_nums = lcg(a, c, m, seed, n)
    sample_mean, sample_std = compute_stats(random_nums)

    # Wartości teoretyczne dla rozkładu jednostajnego na [0,1]
    theoretical_mean = 0.5
    theoretical_std = 1 / np.sqrt(12)

    # Wyświetlenie wyników
    print("Generator LCG:")
    print("Liczba próbek:", n)
    print("Średnia próbki:", sample_mean)
    print("Odchylenie standardowe próbki:", sample_std)
    print("\nWartości teoretyczne dla rozkładu jednostajnego [0,1]:")
    print("Średnia teoretyczna:", theoretical_mean)
    print("Odchylenie standardowe teoretyczne:", theoretical_std)

    # Rysowanie histogramu
    plt.figure(figsize=(8, 6))
    plt.hist(random_nums, bins=50, density=True, alpha=0.75, color='skyblue', edgecolor='black')
    plt.title("Histogram wygenerowanych liczb (LCG)")
    plt.xlabel("Wartości")
    plt.ylabel("Gęstość prawdopodobieństwa")
    plt.grid(True)
    plt.show()

    report = f"""
Sprawozdanie:

Generator LCG (Linear Congruential Generator) został użyty do wygenerowania {n} liczb losowych, operuje on na zasadzie rekurencyjnej zależności:
    X(n+1) = (a * X(n) + c) mod m,
gdzie wygenerowane liczby są następnie normalizowane do przedziału [0,1].

Parametry wygenerowanych liczb:
    Średnia próbki: {sample_mean:.5f}
    Odchylenie standardowe próbki: {sample_std:.5f}

Wartości teoretyczne dla rozkładu jednostajnego na [0,1]:
    Średnia teoretyczna: {theoretical_mean}
    Odchylenie standardowe teoretyczne: {theoretical_std:.5f}

Histogram przedstawia równomierny rozkład wygenerowanych liczb, co potwierdza, że generator dobrze aproksymuje teoretyczny rozkład jednostajny.
Niewielkie odchylenia między wartościami teoretycznymi a próbkowanymi są naturalne przy skończonej liczbie generowanych próbek.

Małe różnice między wartościami teoretycznymi a empirycznymi wynikają z losowego charakteru próbek oraz skończonej liczby generacji.

Histogram potwierdza, że liczby są równomiernie rozłożone, co jest pożądanym efektem dla tego typu generatora.
"""
    print(report)
```
